```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn import svm
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(font_scale=1.2)
```

```python
recipes=pd.read_csv(r'C:\Users\user\OneDrive\桌面\Muffin.csv')
recipes.head()
```

```python
#Create scatterplot of dataframe
sns.lmplot(x='Sugar',y='Butter', data=recipes, hue='Type',
          palette='Set1', fit_reg=False, scatter_kws={"s":70})
```

```python
# Specify inputs for the model
# sugar_butter=recipes[['Sugar','Butter']].as_matrix() 舊版
sugar_butter=recipes[['Sugar','Butter']].to_numpy()
type_label=np.where(recipes['Type']=='Muffin', 0, 1)
```

```python
model=svm.SVC(kernel='linear', C=2**-5)
```

```python
model.fit(sugar_butter, type_label)
```

```python
w=model.coef_[0]
a=-w[0]/w[1]
xx=np.linspace(5,30)
yy=a*xx-(model.intercept_[0]/w[1])
```

```python
b=model.support_vectors_[0]
yy_down=a*xx+(b[1]-a*b[0])
b=model.support_vectors_[-1]
yy_up=a*xx+(b[1]-a*b[0])
```

```python
w=model.coef_[0]
a=-w[0]/w[1]
xx=np.linspace(5,30)
yy=a*xx-(model.intercept_[0]/w[1])
```

```python
b=model.support_vectors_[0]
yy_down=a*xx+(b[1]-a*b[0])
b=model.support_vectors_[-1]
yy_up=a*xx+(b[1]-a*b[0])
```

```python
#Look at the margins and support vectors
sns.lmplot(x='Sugar',y='Butter', data=recipes, hue='Type',
          palette='Set1', fit_reg=False, scatter_kws={"s":70})
plt.plot(xx, yy, linewidth=2, color='black')
plt.plot(xx, yy_down, 'k--')
plt.plot(xx, yy_up, 'k--')
plt.scatter(model.support_vectors_[:, 0], model.support_vectors_[:, 1],
           s=80, facecolors='none')
```

```python
#Plot the hyperplane
sns.lmplot(x='Sugar',y='Butter', data=recipes, hue='Type',
          palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black')
```

```python
#Plot the point to visually see where the point lies
sns.lmplot(x='Sugar',y='Butter', data=recipes, hue='Type',
          palette='Set1', fit_reg=False, scatter_kws={"s": 70})
plt.plot(xx, yy, linewidth=2, color='black')
plt.plot(12,12, 'yo', markersize='9')
```

```python
#Create a function to guess when a recipe is a muffin or a cupcake using the SVM model we created
def muffin_or_cupcake(butter, sugar):
    if(model.predict([[butter, sugar]]))==0:
        print('You\'re looking at a muffin recipe!')
    else:
        print('You\'re looking at a cupcake recipe!')
```

```python
muffin_or_cupcake(12,12)
```
